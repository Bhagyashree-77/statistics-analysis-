# Z SCORE INDICATES HOW MANY STANDARD DEVIATION ON AWAY A DATA POINT IS 

```python
df['zscore']=(df.Height-df.Height.mean())/df.Height.std()
```

```python
df.sample(5)
```

```python
df[df['zscore']>3]
```

```python
df[df['zscore']<-3]
```

# REMOVE THE OUTLIER AND PRODUCE NEW DATAFRAME

```python
df_no_outlier=df[(df.zscore>-3)&(df.zscore<3)]
df_no_outlier
```
